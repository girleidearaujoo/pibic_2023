```python
import pandas as pd
import plotly.express as px
import geopandas as gpd
```

```python
df = pd.read_csv('./san-francisco-crimes-2008-2022_2.csv')
```

```python
from shapely import wkt
df['geometry'] = df['geometry'].apply(wkt.loads)
```

```python
df = df[df.Category.isin(['ASSAULT'])]
df = df.loc[df['Year'].isin([2017])]
```

```python
fig = px.density_mapbox(df, lat='Latitude', lon='Longitude', radius=7, center=dict(lat=df.Latitude.mean(), 
                                lon=df.Longitude.mean()), zoom=11.5,
                        mapbox_style="stamen-terrain", height=900, opacity=0.8)
fig.show()
```
