```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv('./chicago-crimes-2016-2018.csv')
```

```python
df = df[df.Category.isin(['ASSAULT'])]
df = df.loc[df['Year'].isin([2017])]
```

```python
fig = px.density_mapbox(df, lat='Latitude', lon='Longitude', radius=7, center=dict(lat=df.Latitude.mean(), 
                                lon=df.Longitude.mean()), zoom=10.5,
                        mapbox_style="stamen-terrain", height=900, opacity=0.8)
fig.show()
```
