```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('./chicago-crimes-2016-2018.csv')
```

```python
df = df.loc[df['Year'].isin([2017])]
df = df.loc[df['Category'].isin(['ASSAULT'])]
```

```python
df.sort_values('Month', inplace=True)
```

```python
df
```

```python
fig = px.scatter_geo(df, lat='Latitude', lon='Longitude', animation_frame="Month", color_discrete_sequence=["darkblue"])
fig.update_layout(mapbox_style="open-street-map")
fig.update_geos(fitbounds="locations")

fig.show()
```
